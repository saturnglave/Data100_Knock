# １章 ウェブの注文数を分析する１０本ノック

ここでは、ある企業のECサイトでの商品の注文数の推移を分析していきます。  
データの属性を理解し、分析をするためにデータを加工した後、  
データの可視化を行うことで問題を発見していくプロセスを学びます。

### ノック１：データを読み込んでみよう

```python
import pandas as pd
customer = pd.read_csv('customer_master.csv')
customer.head()
```

```python
item = pd.read_csv('item_master.csv')
item.head()
```

```python
transaction1 = pd.read_csv('transaction_1.csv')
transaction1.head()
```

```python
transaction2 = pd.read_csv('transaction_2.csv')
transaction2.head()
```

```python
transaction_d1 = pd.read_csv('transaction_detail_1.csv')
transaction_d1.head()
```

```python
transaction_d2 = pd.read_csv('transaction_detail_2.csv')
transaction_d2.head()
```

### ノック２：データを結合(ユニオン)してみよう

```python
# transaction1 と transaction2 をカラム名継承したままUNIONする
transaction = pd.concat([transaction1, transaction2], ignore_index=False)
transaction.head()
```

```python
# transaction detailのUNION
transaction_d = pd.concat([transaction_d1, transaction_d2], ignore_index=False)
transaction_d.head()
```

### ノック３：売上データ同士を結合(ジョイン)してみよう

```python
# transactionとtransaction_detailをJOINする　価格は除外
```

```python
df = pd.merge(transaction_d, transaction[['transaction_id', 'payment_date', 'customer_id']],
                           on='transaction_id', how='left')
```

```python
df.head()
```

### ノック４：マスタデータを結合(ジョイン)してみよう

```python
# transaction(JOIN済み)に顧客マスタと商品マスタをJOIN
```

```python
df = pd.merge(df, customer, on='customer_id', how='left')
```

```python
df = pd.merge(df, item, on='item_id', how='left')
df.head()
```

### ノック5：必要なデータ列を作ろう

```python
# 売上列を作る
df['price'] = df['item_price'] * df['quantity']
df.head()
```

```python
df[['item_price', 'quantity','price']].head()
```

### ノック6：データ検算をしよう

```python
# 加工前と加工後の売上合計を比較
```

```python
print('加工後', df['price'].sum())
print('加工前', transaction['price'].sum())
```

### ノック7：各種統計量を把握しよう

```python
# 欠損値の有無を確認
df.isnull().sum()
```

```python
# 基本統計量
df.describe()
```

```python
# データの期間を確認
print(df['payment_date'].min(), df['payment_date'].max())
```

### ノック8：月別でデータを集計してみよう

```python
# カラムのデータ型を確認
df.dtypes
```

```python
# 支払日と登録日をdatetime型にする
df['payment_date'] = pd.to_datetime(df['payment_date'])
df['registration_date'] = pd.to_datetime(df['registration_date'])
df.dtypes
```

```python
# 支払月をdatetime型に変換
df['payment_month'] = df['payment_date'].dt.strftime('%Y%m')
df[['payment_date', 'payment_month']].head()
```

```python
# 月別の売上を集計(price列だけ表示)
df.groupby('payment_month').sum()['price']
```

### ノック9：月別、商品別でデータを集計してみよう

```python
# 月別・商品別に売上額の合計と数量を集計
df.groupby(['payment_month', 'item_name']).sum()[['price', 'quantity']]
```

```python
# pivot tableを使って集計
pd.pivot_table(df, index='item_name', columns='payment_month',
               values=['price', 'quantity'], aggfunc='sum')
```

### ノック10：商品別の売上推移を可視化してみよう

```python
# グラフ用のデータを準備
graph_data = pd.pivot_table(df, index='payment_month', columns='item_name', values='price', aggfunc='sum')
graph_data.head()
```

```python
# 可視化
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(list(graph_data.index), graph_data['PC-A'], label='PC-A')
plt.plot(list(graph_data.index), graph_data['PC-B'], label='PC-B')
plt.plot(list(graph_data.index), graph_data['PC-C'], label='PC-C')
plt.plot(list(graph_data.index), graph_data['PC-D'], label='PC-D')
plt.plot(list(graph_data.index), graph_data['PC-E'], label='PC-E')
plt.legend()
```
